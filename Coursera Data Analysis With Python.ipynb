# Data Cleaning

```python
ds.head(4)
```

## Checking The Null Values

```python
ds.isnull()
```

```python
ds.isnull().sum()
```

## Checking The Null Values Through Graph

```python
sc.heatmap(ds.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

## By Handling The Null Values

```python
dsc=ds.fillna(0)
```

```python
dsc.isnull().sum()
```

```python
sc.heatmap(dsc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

## Now Checked The Duplicates Values In DataSets

```python
dsc.duplicated().sum()
```

```python
dub_count=dsc.duplicated(keep=False)
```

```python
heatmap_data=dub_count.astype(int).values.reshape(-1,1)
```

```python
sc.heatmap(heatmap_data,cbar=False,cmap='viridis',yticklabels=False,xticklabels=False)
```

## By Deleting The Duplicates Values

```python
fdsc=dsc.drop_duplicates()
```

## After Removing Duplicates Values Now Checked

```python
fdsc.duplicated().sum()
```

```python
dub_count=fdsc.duplicated(keep=False)
```

```python
heatmap_data=dub_count.astype(int).values.reshape(-1,1)
```

```python
sc.heatmap(heatmap_data,cbar=False,cmap='viridis',yticklabels=False,xticklabels=False)
```

## Drop The Unnecessary Columns

```python
fd=fdsc.drop(['Review','Duration'],axis=1)
```

```python
fd.columns
```

## Handle The Unnecessary Values In Data

```python
fd1=fd.replace(['[]'],'No Specified')
```

```python
fd2=fd1.rename(columns={'Duration.1':'Duration to complete'})
```

```python
fd2.columns
```

```python
fd2['Level']=fd2['Level'].replace(0,'No Specified')
```

```python
fd2['Schedule']=fd2['Schedule'].replace(0,'Fixed Schedule')
```

# Data Analysis & Visualization

```python
fd2.head()
```

## Checking The Courses By Level

```python
fd2['Level'].value_counts()
```

```python
fd2['Level'].value_counts().plot(kind='bar',color=['y','b','gray','black'],width=0.8)
```

## The most popular courses based on their reviews

```python
fd2['Number Of Review']=pd.to_numeric(fd2['Number Of Review'],errors='coerce')
```

```python
sort_data=fd2.sort_values(by='Number Of Review', ascending=False)
```

```python
top_review=sort_data[['Course Title','Number Of Review']].head(10)
```

```python
top_review
```

```python
 colors=['y','b','red','k','gray','orange','purple','green']
 sc.barplot(data=top_review,x='Number Of Review',y='Course Title',palette=colors,hue='Course Title')
 mt.title("Top Courses On The Basis Of Reviews")
```

## Instructor Impact On Course Success

```python
top_intructors=sort_data[['Instructor','Number Of Review','Course Title']].head(5)
```

```python
top_intructors
```

```python
sc.barplot(data=top_intructors,x='Number Of Review',y='Instructor',hue='Course Title')
mt.legend(loc='lower right')
```
